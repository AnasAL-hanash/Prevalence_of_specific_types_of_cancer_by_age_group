Prevalence of specific types of cancer by age group

Analyzing a dataset related to specific types of cancer (prevalence of different types of cancer in different age groups) and using a graph using matplotlib library helps in understanding the data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

 Generate data


```python
# Generate data using numpy
# Age groups
ages = np.array(['0-18', '19-35', '36-50', '51-65', '65+'])
```

```python
# Rates of different types of cancer for each age group (default values)
cancer_rates = np.array([
    [0.1, 0.2, 0.1, 0.3, 0.05],  # breast cancer
    [0.05, 0.15, 0.2, 0.25, 0.1],  #Colon cancer
    [0.02, 0.05, 0.1, 0.2, 0.3],  # Lung cancer
])
```

```python
cancer_types = ['Breast Cancer', 'Colon Cancer', 'Lung Cancer']
```

Organize data using pandas

```python
df = pd.DataFrame(cancer_rates.T, columns=cancer_types, index=ages)
print("DataFrame:\n", df)
```

Plotting data using matplotlib

```python
df.plot(kind='bar', figsize=(10, 6))
plt.title('Cancer Rates by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Cancer Rate')
plt.xticks(rotation=0)
plt.legend(title='Cancer Type')
plt.show()
```
